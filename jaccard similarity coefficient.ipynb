```python
# jaccard accuracy for cnn model, crnnmodel, and cGrnn model on MTAT test set

import keras
# import tensorflow
import numpy as np
from os import listdir

max_jcc = 0
idx = -1
path = 'C:/DT'
x_test = np.load(path+'\\x_test_MelSpec.npy')
y_test = np.load(path+'\\y_test.npy')

for i, file in enumerate(listdir('C:/DT/model_save/crnn/')):
#     model = keras.models.load_model('C:/DT/model_save/cnn/'+file)

    model = keras.models.load_model('C:/DT/model_save/crnn/'+file)

#     model = keras.models.load_model('C:/DT/model_save/cGrnn/'+file,custom_objects={'tf':tensorflow})

    y_pred = model.predict(x_test)
    y_pred = np.where(y_pred > 0.5, 1, 0)
    from sklearn.metrics import jaccard_similarity_score
    jcc = jaccard_similarity_score(y_test, y_pred)
    print(jcc)
    if max_jcc < jcc:
        max_jcc = jcc
        idx = i+1
print('MAX',max_jcc,idx)
```
